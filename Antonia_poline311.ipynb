```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
pd.set_option('display.float_format', lambda x: f"'{x:.2f}")

sns.set(style='whitegrid')
```

```python

df=pd.read_csv('https://raw.githubusercontent.com/poline822-arch/healthcare-stroke-synthetic.csv/refs/heads/main/311_Service_Requests_from_2019May%20(1).csv')

df.head()
```

```python
#Find the total number of rows and columns and their data types

print('shape(row,colunm):',df.shape)
print('data types:',df.dtypes)
df.info()
```

```python
 #Find the summary statistics for the file

 df.describe(include='all')

```

```python
#Analyze the data and find how many missing values are in the dataset
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.isna().sum().sum()
```

```python
#Find missing values in all columns

df.isnull().sum()
```

```python
#Use imputation techniques( mean, median etc.) to clean the data

df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors='coerce')
df['Due Date'] = pd.to_datetime(df['Due Date'], errors='coerce')
print(df)

```

```python
#9. Verify if the new data is now fully clean

df.isnull().sum()
```

```python
#. Create atleast 2 to 3 visualizations

sns.histplot(df['Borough'], kde=True)
plt.title('Distribution of Borough')
plt.xlabel('Borough')
plt.ylabel('Frequency')
plt.show()
```

```python
#. Create atleast 2 to 3 visualizations
df['Incident Zip'] = pd.to_numeric(df['Incident Zip'], errors='coerce')
sns.histplot(df['Incident Zip'].dropna(), kde=True)
plt.title('Distribution of Incident Zip')
plt.xlabel('Incident Zip')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Location Type', y='Incident Zip', data=df)
plt.title('Location Type vs. Incident Zip')
plt.xlabel('Location Type')
plt.ylabel('Incident Zip')
plt.show()
```
