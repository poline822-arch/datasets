```python
import pandas as pd

url = "https://raw.githubusercontent.com/reachingneeraj/TestDatasets/refs/heads/main/Student_Enrollment_2010-11_School_Year.csv"
df = pd.read_csv(url)

df.head()
df.tail()
```

```python
# 3. Shape
df.shape


```

```python
# 4. Data types
df.dtypes

```

```python
# 5. Summary
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Option 2: Fill categorical with 'Unknown'
df.fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
df.shape
```

```python
sns.histplot(df['All Students'], bins=30, kde=True)
plt.title("Distribution of Enrollment Count")
plt.show()
```

```python
sns.boxplot(x=df['All Students'])
plt.title("Boxplot of Enrollment Count")
plt.show()
```

```python
df['GradeLevel'].value_counts().plot(kind='bar')
plt.title("Distribution by Grade")
plt.show()
```

```python
gender_counts = pd.Series({
    'Female': df['Female'].sum(),
    'Male': df['Male'].sum(),
    'Gender X': df['Gender X'].sum()
})
gender_counts.plot(kind='bar')
plt.title("Distribution by Gender")
plt.ylabel("Total Students")
plt.xlabel("Gender")
plt.show()
```
